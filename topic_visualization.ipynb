# Loading the models for model visualization

```python
from gensim.models import LdaModel, Phrases
from gensim.corpora import Dictionary

import pyLDAvis.gensim
pyLDAvis.enable_notebook()

from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer

import load_lee_background_corpus as load_texts
```

### Loading the previously saved models

```python
dictionary = Dictionary.load('dicionario_1')
phraser = Phrases.load('frases_1')
model = LdaModel.load('modelo_1')

train_texts = load_texts.get_train_texts()
corpus = [dictionary.doc2bow(text) for text in train_texts]
```

### Display the model's topics

```python
model.top_topics(texts=train_texts, dictionary=dictionary, coherence='c_v')
```

### User pyLDAvis to visualize the topics

The red bars represent the frequency of a term in a given topic, (proportional to p(term | topic)), and the blue bars represent a term's frequency across the entire corpus, (proportional to p(term)). Change the value of λ to adjust the term rankings -- small values of λ (near 0) highlight potentially rare, but exclusive terms for the selected topic, and large values of λ (near 1) highlight frequent, but not necessarily exclusive, terms for the selected topic.

```python
pyLDAvis.gensim.prepare(model, corpus, dictionary)
```
